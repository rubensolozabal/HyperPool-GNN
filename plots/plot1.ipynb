```python
import pandas as pd
import json
```

```python
# Define the directory containing your JSON files
directory = 'results/plot1/'

# Get a list of all files in the directory
all_files = os.listdir(directory)

# Filter out only JSON files
json_files = [file for file in all_files if file.endswith('.json')]

# Initialize an empty list to store DataFrames
data_frames = []

# Loop through the JSON files and append their contents to the list
for file in json_files:
    file_path = os.path.join(directory, file)
    with open(file_path) as f:
        data = json.load(f)
        data_frames.append(pd.DataFrame([data]))

# Concatenate all data frames into a single DataFrame
df = pd.concat(data_frames, ignore_index=True)
```

```python
# Extract the following keys: 'dataset', 'model', 'auc', 'auc_std'
df_plot = df[['dataset', 'model', 'acc', 'acc_std', 'alpha']]
# df_cora = df_plot[df_plot['dataset']=='Cora']


# Change the value alpha=0.0 to alpha=1E-5
df_plot = df_plot.replace({'alpha': {0.0: 1E-5}})
```

```python
# Substact the minumum value for acc groped by dataset
df_plot['acc'] = df_plot.groupby('dataset')['acc'].transform(lambda x: x - x.min())
```

```python
# Mutliply acc by 100 and acc_std by 100
df_plot['acc'] = df_plot['acc'] * 100
df_plot['acc_std'] = df_plot['acc_std'] * 100
```

```python
df_plot
```

```python
# Plot using seaborn, x axis is alpha, y axis is auc, hue is model
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

palette = ["#F72585", "#7209B7", "#3A0CA3", "#4361EE", "#4CC9F0"]

fig, ax = plt.subplots(figsize=(5, 4))

# Plot the responses for different events and regions
sns.lineplot(
    x="alpha", 
    y="acc",
    hue="dataset",
    data=df_plot,
    ax=ax, 
    palette=palette, 
    linewidth=1.5,
    # err_style="band",  # This will display the standard deviation as a band
    # errorbar="sd"  # This assumes your DataFrame contains standard deviation information
)

# Loop through each dataset to add standard deviation as shaded area
for name, group in df_plot.groupby('dataset'):
    # Sort values by 'alpha' to ensure correct filling
    group = group.sort_values(by='alpha')
    ax.fill_between(
        group['alpha'], 
        group['acc'] - group['acc_std'],  # Lower bound
        group['acc'] + group['acc_std'],  # Upper bound
        color=group['dataset'].map({'Cora': '#3A0CA3', 'CiteSeer': '#7209B7', 'PubMed': '#F72585'}),
        alpha=0.2  # Adjust transparency of the shaded area
    )
# Set x axis limits log scale
plt.xlim(1E-5, 200) 
plt.ylim([-1, 2])

plt.xlabel(r'Uniformity regularization $\alpha$', fontsize=16)
plt.ylabel(r'$\Delta$ Accuracy (%)', fontsize=16)

handles, labels = ax.get_legend_handles_labels()
# labels=['b=1,d=3', 'b=1,d=5', 'b=1,d=10', 'b=2,d=3']
# handles[3].set_linestyle("--")
ax.legend(title=None, loc='lower right', handles=handles, labels=labels)

ax.set_xscale('log')
plt.grid(True, which="both", ls="-")

# plt.savefig("plots/A1.pdf", bbox_inches='tight')
plt.show()
```
